###**<code>A- </code>**

**1.** Create an ndarray, named `M` , of size `4×4` containing integers from 0 to 15 . Display `M`.



```python
M = np.arange(16).reshape(4,4)
M
```

**2.** Display the 2nd row of `M` , then display its last element. 


```python
print(M[1])
print(M[3][3])
```

**3.** Display the elements of the diagonal of `M`.

```python
for i in range(4):
  print(M[i][i])
```

**4.** Create an array `ndarray` `M1` that represents the result of transposing the last two columns of `M`. 


```python
A = M[:,2:4]
M1 = A.transpose()
print(M1)
```

**5.** Swap the two rows of `M1` . Display `M1` 

```python
M1[[0,1]] = M1[[1,0]]
print(M1)
```

###**<code>B- </code>**


**1.** Create an array of dimension 1 , named `vect` , with 12 random elements between 0 and 20 . Display `vect`.


```python
vect = np.random.randint(20, size=(12))
print(vect)
```

**2.** Transform `vect` into a two-dimensional (2D) array, named `t_b` , of 4 rows and 3 columns. Display `t_b`.


```python
t_b = np.array(vect).reshape(4, 3)
print(t_b)
```

**3.** Display the size, dimension and number of elements of `t_b`.
 

```python
print(np.shape(t_b))
```

**4.** Display the maximum of each row of `t_b`.


```python
for i in range(4):
  print("max of row",i+1,"is",np.max(t_b[i]))
```

**5.** Display the sum of each column of `t_b`.


```python
for i in range(3):
  print(np.sum(t_b[:,i]))
```

**6.** Transform `t_b` into an array of size 4×4 , named `t_a` . Display `t_a`.


```python
t_a = np.resize(t_b,16).reshape(4, 4)
print(t_a)
```

**7.** Decompose `t_a` vertically into 4 arrays `X` , `Y` , `Z` and `T` with 4 elements each.


```python
X = t_a[:,0]
Y = t_a[:,1]
Z = t_a[:,2]
T = t_a[:,3]
print(X,Y,Z,T)
```

**8.** Merge the array `X` and `t_a` so that the rows of `X` are added as a new column to `t_a`.


```python
t_a = np.hstack((t_a,X.reshape((4,1))))
print(t_a)
```

**9.** Define an array `t_u` of dimension 1 containing the elements of `t_a` sorted in ascending order so that no element is duplicated. Display `t_u`.


```python

t_u = [t_a[i][j] for i in range(len(t_a)) for j in range(len(t_a[0]))]
t_u.sort()
print(t_u)
```

**10.** Transform `t_u` into an array of size `4×3` . Then insert the elements of `Y` in the second column of `t_u` . Display `t_u`.


```python
t_u = np.resize(t_u,12).reshape(4,3)
t_u = np.hstack((t_u,Y.reshape(4,1)))
print(t_u)
```
